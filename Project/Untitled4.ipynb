```python
# 3. AGENT SYSTEM INSTRUCTIONS
instructions = "You are an elite Crime Analyst. Provide logical, sharp, and professional detective analysis."
model = genai.GenerativeModel(model_name='gemini-2.5-flash', system_instruction=instructions)

def run_crime_agent():
    print("🕵️‍♂️ --- CRIME INVESTIGATION TERMINAL --- 🕵️‍♂️")

    while True:
        print("\n--- SELECT A CASE ---")
        for cid, data in crime_cases.items():
            print(f"[{cid}] : {data['name']}")
        print("[exit] : To Close Program")

        user_choice = input("\nEnter ID (e.g. 001) or exit: ").strip().lower()

        if user_choice == 'exit':
            print("🔴 Program Exited. Case Files Closed.")
            break

        if user_choice in crime_cases:
            case = crime_cases[user_choice]
            print(f"\n🔍 ANALYZING: {case['name']}...")

            prompt = f"Case: {case['name']}\nDescription: {case['desc']}\nTask: {case['req']}"

            try:
                response = model.generate_content(prompt)
                print("\n" + "="*60)
                print(f"📝 DETECTIVE REPORT - CASE {user_choice}")
                print("="*60)
                print(response.text)
                print("="*60)
            except Exception as e:
                print(f"Error: {e}")
        else:
            print("⚠️ Invalid Selection! Type a valid ID or 'exit'.")

# RUN
run_crime_agent()
```
